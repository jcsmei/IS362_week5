## **1. What is the northernmost airport in the United States?**

```python
airports['tzone'].unique()
```

As you can see from the unique values on 'tzone' (time zone).

There are none America cities. We will exclude Asia in a new dataframe.

```python
ap = airports[~airports.tzone.str.contains('Asia')]
```

```python
ap.sort_values('lat', ascending=False).head()
```

As you can see from the sorted list above, the two northernmost airports by latitude (the 'lat' column) are:
1. Dillant Hopkins Airport
2. Wiley Post Will Rogers Memorial Airport

A quick search on Google maps returns the following results.
Wiley Post Will Rogers Memorial Airport is the northern most airport that belongs to the US.

**However, Dillant Hopkins Airport is the northernmost airport in the continental US.**

Lets revisit the weather dataframe.

```python
weather.tail()
```

**What is the difference between wind speed and wind speed?**

According to [www.differencebetween.net]('http://www.differencebetween.net/science/nature/difference-between-gust-and-wind/')

`"All gusts are a type of wind. A gust is a sudden increase of the wind's speed that lasts no more than 20 seconds. This usually occurs when wind speeds reach a peak of at least 16 knots. A wind gust usually comes in 2-minute intervals."`

We will not be using wind_gust.

```python
weather['origin'].unique()
```

Here we have three major airports in the New York area:

1. EWR (Newark Liberty International Airport)
2. JFK (John F. Kennedy International Airport)
3. LGA (LaGuardia Airport)

```python
weather.isnull().sum()
```

The output shows the overall NULL (or NaN) values on all columns.

We will focus on wind_speed since wind_gust is mostly NULL and it does not affect analysis.

Lets sort the dataframe in descening order on 'wind_speed', from highgest to lowest.

```python
weather.sort_values('wind_speed', ascending=False).head()
```

Let's be sure and isolate the date of February 12, 2013.

```python
we = weather[weather.time_hour.str.contains('2013-02-12')]
```

```python
we.sort_values('wind_speed', ascending=False).head()
```

As you can see on 2013-02-12 at hour 3, Newark had the windiest weather in the NY area.

[It is historically recorded in www.wunderground.com](https://www.wunderground.com/history/daily/us/nj/newark/KEWR/date/2013-2-12)

![Historic_Data](http://pichost.org/images/2018/09/30/hist.png)

As you can see in 3:51 AM the wind speed is recorded as 1048 mph!

```python
we.groupby('origin').wind_speed.max()
```
